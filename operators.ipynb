```python
#Operators

#1-Arithmetic Operators

#1)Addition(+)
x=3453
y=8896

print("Addition:",x+y)

#2)Subtraction(-)

print("Subtraction:",x-y)

#3)Multiplication(*)

print("Multiplication:",x*y)

#4)Division(/)

print("Division:",x/y)

#5)Modulus(%)

print("Modulus:",x%y)

#6)Exponential(**)
 
print("Exponential:",x**y)

#7)Flow-division(//)

print("Flow-division:",x//y)
```

```python
#Assignment Operators

a=40
a+=10
print(a)

b=55
b+=10
print(b)

c=32
c*=10
print(c)

d=986
d%=9
print(d)

e=86
e**=10
print(e)

f=765 
f//=4
print(f)

g=65 
g/=5
print(g)

h=98
h^=15
print(h)

i=650
i&=12
print(i)
```

```python
#3-Comparission Operators

k=767
p=988

print(k==p) #equal

print(k!=p) #not equal

print(k>p) #greater than

print(k<p) #less than

print(k>=p) #greater than or equal

print(k<=p) #less than or equal
```

```python
#4-Logical Operators

x=10
print(x>8 and x>5) #logical and

print(x>10 or x>5) #logical or
 
print(not(x>10 and x>5)) #logical not
```

```python
#5-Identity Operators

#is

list1=[456,675,353,976]
list2=[567,345,875,323]
x=list1
print(x is list1)
print(list1 is list2)

#is not

print(list1 is not list2)
print(x is not list2)
print(x is not list1)
```

```python
#6-Membership Operators

#in

print(x in list2) #returns false because we don't have x value in list2

print(x in list1)

print(123 in list1)


#not in

print(x not in list1)

print(x not in list2)

print(123 not in list1)
```

```python
#7-Bitiwise Operators

#Bitiwise AND(&)

print(10 & 6)

#Bitiwise OR(|)

print(10 | 6)

#Bitiwise XOR(^)

print(10 ^ 6)

#Bitiwise NOT(~)

print(~10)

#Left shift(<<)

print(10 << 6)

#Right shift(>>)

print(10 >> 6)

```
